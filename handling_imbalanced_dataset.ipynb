**Imbalanced Dataset**

A dataset wiith an unequal class distribution

for example our diabetic dataset which has 1000 datapoint of people with +ve diabetes and 100 with negative ,

we can't feed such data to our ml model,we'll need to process it inorder to remove this imbalance  

```python
#importing the dependencies
import numpy as np
import pandas as pd
```

```python
#loading the dataset to pandas dataframe
credit_card_data = pd.read_csv('/content/credit_data.csv')
```

*This dataset contains many transaction in which some are fraudulant and some are not*

*we will use to check if a transaactio is legit or fraud*




```python
print(credit_card_data.head())
```

**0->>legit transaction**

**1-->fraud**

```python
credit_card_data.tail()
```

```python
#determine the distribution of the 2 classes
credit_card_data['Class'].value_counts()
```

**This is Highly imbalanced dataset**

```python
#separating the legit and fraudulant transactions
legit = credit_card_data[credit_card_data.Class == 0]
#it'll take all the rows which contains the class value as 0
fraud = credit_card_data[credit_card_data.Class == 1]
#this will contain all tthe fraudulant data points
```

```python
print(legit.shape)
print(fraud.shape)
#
```

**UnderSampling:-we'll take random datapoints sample from the legit ones equivalent to the fraud ones**

*building a sample dataset containing similar distribution of legit and fraudulant transaction.*

number of fradulant transaction=165

we don't need to take exact same value we can take round of to it.

```python
legit_sample = legit.sample(n=165)
```

```python
print(legit_sample.shape)
```

```python
#concatinate the 2 dataframes
new_dataset = pd.concat([legit_sample,fraud],axis=0)
```

```python
print(new_dataset.head())
```
