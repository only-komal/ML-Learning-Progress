# `**Feature Extension**`

*The Mapping from textual data to real valued vectors is called feature extraction*

**Bag of Words[BOW]**:- List of unique words in the text corpus

**Term Frequency inverse Document Frequency[TF-DF]**:-To count the number of times each word appears in a document.

# `**TF-DF vectorizer**`

**Term Frequency(Tf)**:-(Number of times term t appears in a document)/(Number of terms in the document)

This can tell us what are the important words are.

**Inverse Document Frequency(IDF)**:-log(N/n),where N is the number of documenting and n is the number of documents a term t has appeared in.

IDF value of a rare word is high ,whereas the IDF of a frequent word is low.


`**TF-IDF** value of a term :- *TF ✖ IDF*`
