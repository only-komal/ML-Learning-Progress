List vs Numpy-Time Taken

```python
np_array=np.array([i for i in range(10000)]) #creating the numpy array
#using numpy abbrevation np.array func
start_time=process_time()
np_array**2 #np array orginal raised to the the power 2
# i.e each element of the existing np_array
end_time=process_time()
print(end_time-start_time)
```

whereas the same task if done using list would've taken more time

```python
python_list=[i for i in range(10000)]
start_time=process_time()
python_list=[i**2 for i in python_list]
end_time=process_time()
print(end_time-start_time)
```

it's not much of difference as its was only for 10,000 value but when we'll work with million of data points it'll be quiet significant.

we use the func ".zeros()"and inside the paranthesis we enter the shape of our arrayi.e the no. of rows and columns

```python
x=np.zeros((4,5))
print(x)
```

2.To create array of ones

use ".ones"

```python
y=np.ones((3,3))

print(y)
```

3.Create an array of particular value


use ".full"

it'll take first value as the dimensions of array and 2nd value as the number with which it has to be initialised

```python
z=np.full((5,4),5)
print(z)
```

```python
a=np.eye(4)
print(a)
```

**Mathematical operations on an array**

the "+" will concatinate the two lists if used simply ,i.e we can't add values in case of lists but we can do it in terms of numpy arrays



```python
list1=[1,2,3,4,5]
list2=[6,7,8,9,10]

print(list1 + list2)
```

```python
a=np.random.randint(0,10,(3,3))
b=np.random.randint(10,20,(3,3))

print(a)
print(b)
```

```python
print(a+b) #print(np.add(a,b))

print(a-b) #print(np.subtract(a,b))

print(a*b) #print(np.multiply(a,b))

print(a/b)#print(np.divide(a,b))
```

**Array Manipulation**

```python
array=np.random.randint(0,10,(2,3))
print(array)
print(array.shape)
```

```python
#to create the transpose of the given array

#transpose=change rows into columns
trans=np.transpose(array)
print(trans)
print(trans.shape)
```

```python
array=np.random.randint(0,10,(2,3))
print(array)
print(array.shape)
```

```python
#we can do the transpose using ".T" as well
trans2=array.T
print(trans2)
print(trans2.shape)
```

```python
#reshaping the array

a=np.random.randint(0,10,(2,3))
print(a)
print(a.shape)
```

```python
b=a.reshape(3,2)
print(b)
print(b.shape)
```
