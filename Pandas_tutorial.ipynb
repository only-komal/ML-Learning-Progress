The head() method in pandas DataFrames is designed to display the first 5 rows by default.

This is a common convention in data analysis to quickly inspect the beginning of a dataset without printing the entire contents, which can be very large.

You can change the number of rows displayed by passing an integer argument to head(), for example, housing_df.head(10) would show the first 10 rows.

**Manipulating a DataFrame**


```python

#adding a column to dataframe
#when we earlier added all the info about data and features of th ehousing data
#but left target value to be added we'll add it now to get it as price
housing_df["price"]=housing_dataset.target
```

```python
#will print the starting values of it again
housing_df.head()
#will be able to see a column named price in it now
```

```python
#removing a particular row
housing_df.drop(index=0,axis=0)
#drop function to remove any row/column
#axis=0 for removing row and axis=1 for removing column
#index is the row you want to
#will be removed temporarily i.e just for this command and will reappear in next
#if want to reomve permanently create another variable and store in that
#housing_def2=housing_def.drop(index=0,axis=0)
```

```python
#drop a column
housing_df.drop(columns='price',axis=1)
```

```python
#locating a row using the index value
housing_df.iloc[0]
#iloc is the function used to locate particular row or column
#will give us all the values of mention index
```

```python
#locating a particular column
print(housing_df.iloc[:,0])
# 0th column will be located i.e it's value will be printed
housing_df.iloc[:,-1] #will give value of last column
```

**Co-relation**

1.positive:-values in one column increases when the values in other particular column increases

2.negative:-values in one column decreases when the values in other particular column increases

co-realtion is basically finiding relation betweeen all the available columns



```python
#will give the corelation values ,-ve values means they have -ve corelation
# +ve values means they have positive co-relation
housing_df.corr()
#each column will be compared with every other column out there

```
