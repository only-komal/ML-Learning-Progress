Importing libraries

```python
 import seaborn as sns
 import matplotlib.pyplot as plt
 import numpy as np
 import pandas as pd
```

**Seaborn has some built-in datasets**

```python
#for much detailed explanations of any particular explorations check seaborn documentation from chrome
#total bill vs tip dataset
tips = sns.load_dataset('tips')
#this will be imported as pandas dataframe initial values of which can be printed using head function
tips.head()
```

```python
#visualise the data
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
#data=name the data we want to plot
#x-axis=name of the column we  want to have on x-axis
#y=name of the column we want on y-axis
#col=basis of plots (here we'll have 2 based on dinner and sun i.e time)
#hue,style,size=differentiators of the plot
```

In matplotlib we manually have to set these color and size things but in seaborn we need not to

WE CAN SET DEFAULT THEME FOR THE PLOTS AS WELL

```python
#setting a  theme for the plots
sns.set_theme()
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

It's more spacious as well as we got bricks  

Once sns.set_theme has run in the code you need not to run it everytime

```python
#housin price dataset
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
house=pd.DataFrame(housing.data,columns=housing.feature_names)
house['price']=housing.target
house.head()
```

Distribution plot

```python
#tells in what range max value lies
sns.distplot(house['price'])
#will give us the probablity curve
```

Correlation

-positive

-negative



Heat Map

```python
#first we need to create this co-relation
corelation=house.corr()
```

```python
#constructing gga heat map
plt.figure(figsize=(10,10))
sns.heatmap(corelation,cbar=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
#
```
