```python
#importing matplotkib.py
import matplotlib.pyplot as plt
```

To plot we need some data first


```python
#import numpy to get data for our plots
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

If we plot the x and y we'll get asin curve and

if we plot x and z we'll get a cos curve

```python
#sin wave
plt.plot(x,y) #creates the wave
plt.xlabel('x') #give label to x axis
plt.ylabel('sin(x)') #give label to y axis
plt.title('sin wave') #give title to the graph
plt.show() #to print the plotted graph
```

```python
#cosine wave
plt.plot(x,z)
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.title('cosine wave')
plt.show()
```

```python
#create a parabola
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('parabola')
plt.show()
```

```python
#plot with different symbols not a simple line

plt.plot(x,y,'r+') #r=red colour of the graph
# + = replace line + symbols
plt.xlabel('x')
plt.ylabel('y')
plt.title('parabola')
plt.show()
```

```python
#can do with other colours as well as symbols too
plt.plot(x,y,'g.')
plt.xlabel('x')
plt.ylabel('y')
plt.title('parabola')
plt.show()
```

Search on google to matplotlib documentation to see what all different colours and symbols you can use


```python
#how we can draw multiple lines/symbols in single graph
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-') #sin graph  between x and sin(x) in green colour with hyphen
plt.plot(x,np.cos(x),'r--') #cos graph between x and cos(x) in red colour with double hyphen
plt.show()
```

**Bar Plot**


```python
#creatibg bar graph
fig=plt.figure()  #will create an empty plot
ax=fig.add_axes([0,0,1,1]) # 0,0 represents origin ,1,1 represents height and width
#going to create an rectangle empty figure
languages=['English','French','Spanish','Latin','German'] #creating list
people=[100,50,150,40,70]
ax.bar(languages,people)#to make a bar graph
plt.xlabel('Languages')
plt.ylabel('Number of speakers')
plt.title('Bar Graph')
plt.show()
```

**Pie Chart**

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['English','French','Spanish','Latin','German']
people=[100,50,150,40,70]
ax.pie(people,labels=languages,autopct='%1.2f%%')#pie chart of no.of people but labels must be languages
#autopct=how many decimal/floating points we need
```

**Scatter Plot**

```python
x=np.linspace(0,10,30)
y=np.sin(x)
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('scatter plot')
plt.show()
```

**3-D Scatter plot**

```python

fig3=plt.figure()
ax=plt.axes(projection='3d')#rectangle was 2-d ,here we need 3-d
z=20 * np.random.rand(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
```
